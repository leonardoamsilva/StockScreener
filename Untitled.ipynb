```python
ticker = 'petr4'
url = f'https://statusinvest.com.br/acoes/{ticker}'
headers = {'User-Agent': 'Mozilla/5.0'}

response = requests.get(url, headers=headers)
print(f"Status da Requisição: {response.status_code}") # Deve retornar 200
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
print("Objeto Soup criado. Pronto para a investigação!")
```

```python
# Comece BEM de cima. Vamos pegar o corpo inteiro da página.
body = soup.find('body')

# Agora vamos "printar" essa primeira camada para ver o que tem dentro.
# .prettify() ajuda a formatar o HTML para ficar mais legível.
print(body.prettify())
```
